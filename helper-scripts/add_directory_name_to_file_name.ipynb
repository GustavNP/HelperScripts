```python
import os
import shutil
```

```python
def add_directory_name_prefix_to_file_name(input_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            old_full_file_path = os.path.join(root, file)
            filename_with_dir_prefix = root.split("\\")[-1] + '-' + file
            new_full_file_path = os.path.join(root, filename_with_dir_prefix)
            
            # If the file already exists in the directory, add number to make it unique
            final_filename = filename_with_dir_prefix
            counter = 1
            while os.path.exists(new_full_file_path):
                file_name, file_extension = os.path.splitext(filename_with_dir_prefix)
                final_filename = f"{file_name}_{counter}{file_extension}"
                new_full_file_path = os.path.join(root, final_filename)
                counter += 1

            print(old_full_file_path)
            print(new_full_file_path)
            os.rename(old_full_file_path, new_full_file_path)
```

```python

# Input and output directory paths
input_dir = "C:\\Users\\admin\\Downloads\\archive1\\val"

# Call the function to copy files
add_directory_name_prefix_to_file_name(input_dir)
```
